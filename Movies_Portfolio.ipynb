```python
#Importing needed libraries for analysis

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12, 8) #Adjusts the config of the plots being created
```

```python
#Reading in the data
datasource = r"E:/Data/Movies/movies.csv" 

df = pd.read_csv(datasource)
```

```python
#Taking a look at the data
df.head()
```

```python
#Checking for missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#Dropping rows with missing data 
df = df.dropna()
df.head()
```

```python
#Checking data types of columns

df.dtypes
```

```python
#Cleaning up floats into integers
df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
#Ordering the data
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#Drop Duplicates
df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
df.drop_duplicates()
```

```python
df.head()
```

```python
#Building a scatterplot comparing budget with gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget For Film')

plt.show()
```

```python
#Plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color':'red'}, line_kws={'color':'blue'})
```

```python
#Look for numerical correlation within the dataset
#First I need to create a df of only the numerical columns
dfnum = df[['budget', 'gross', 'runtime', 'score', 'votes', 'year']]
```

```python
dfnum.head()
```

```python
df.head()
```

```python
dfnum.corr() #pearson, kendall, spearman
correlation_matrix = dfnum.corr(method='pearson')
```

```python
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
#Looking for correlations in Company
dfcopy = df
for col_name in dfcopy.columns:
    if(dfcopy[col_name].dtype == 'object'):
        dfcopy[col_name] = dfcopy[col_name].astype('category')
        dfcopy[col_name] = dfcopy[col_name].cat.codes
        
dfcopy
```

```python
df_numerized = dfcopy
```

```python
correlation_matrix2 = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix2, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
corr_pairs = correlation_matrix2.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
